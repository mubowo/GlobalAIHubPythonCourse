Assignment :
* Create three classes named Animals, Dogs and Cats. Add some features to these classes. Create some functions with these attributes. Don't Forget! you have to do it using inheritance.

```
class Animals:
  def __init__(self, fname, lname, lives):
    self.fname = fname
    self.lname = lname
    self.lives = lives

  def print_name(self):
    print(self.fname + " " + self.lname)

  def print_lives(self):
    print(self.fname + " " + self.lname + " lives in " + self.lives)

class Dogs(Animals):
  def __init__(self, fname, lname, lives, eats, sleeps):
    super().__init__(fname, lname, lives)
    self.eats = eats
    self.sleeps = sleeps

  def print_dogs_descriptions(self):
    print(self.fname + " " + self.lname + " dog is lives in " + self.lives + ", eating " + self.eats + " and sleeps " + self.sleeps)

class Cats(Animals):
  def __init__(self, fname, lname, lives, eats, sleeps, toys):
    super().__init__(fname, lname, lives)
    self.eats = eats
    self.sleeps = sleeps
    self.toys = toys

  def print_cats_descriptions(self):
    print(self.fname, self.lname, " cat is lives in ", self.lives, ", eating ", self.eats, " and sleeps ", self.sleeps, " playing with ", self.toys)
```

```
dog_1 = Dogs("Welsh", "Corgi", "DogHouse", "Dog Food", "At night")
dog_1.print_dogs_descriptions()
```

```
cat_1 = Cats("American", "Curl", "Cattery House", "Cat Food", "For 12 Hours a day", "Mini Tennis Ball")
cat_1.print_cats_descriptions()
```
