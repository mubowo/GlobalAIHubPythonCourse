### Final Project

Description:
* Company Management System, Determine the name and ages of employees, company managers, and the languages they can speak. Then write a program that will print the languages that any of the employees can speak. In this project, complete the project by creating two classes named employees and manager.

```
class People:
  def __init__(self, fname, lname, age):
    self.fname = fname
    self.lname = lname
    self.age = age
    
class Employees(People):
  def __init__(self, fname, lname, age):
    super().__init__(fname, lname, age)
    self.mgmt_type = "Employee"
    self.language = set() # sets of new language

  def inputLanguage(self, additional_lang):
    self.language.add(additional_lang)
    print("Success insert additional language", additional_lang)
  
  def print_employees_info(self):
    print(self.fname, self.lname, " is an ", self.mgmt_type, " with ", self.age, "years old age.")
    print("He/She can speak: ")
    for i in self.language:
      print(i)


class Manager(People):
  def __init__(self, fname, lname, age):
    super().__init__(fname, lname, age)
    self.mgmt_type = "Manager"
    self.language = set() # sets of new language

  def inputLanguage(self, additional_lang):
    self.language.add(additional_lang)
    print("Success insert additional language", additional_lang)

  def print_manager_info(self):
    print("{} {} is a {} with {} years old age.".format(self.fname, self.lname, self.mgmt_type, self.age))
    print("He/She can speak: ")
    for j in self.language:
      print(j)
```
